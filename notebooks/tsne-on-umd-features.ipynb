```python
import numpy as np
import pandas as pd
from sklearn import manifold
import matplotlib.pyplot as plt
```

```python
# df_train = pd.read_csv("../features/train_features.csv")
# df_test =pd.read_csv("../features/test_features.csv")

df_train = pd.read_csv("../features/train_ae_features.csv")
df_test =pd.read_csv("../features/test_ae_features.csv")

print(f"train shape is {df_train.shape}, test shape is {df_test.shape}")
```

### Mixup

```python
df_mixup =pd.read_csv("../features/mixup_ae_features.csv")
df = pd.concat([df_train, df_test, df_mixup], axis=0)
use_col = [col for col in df.columns if col != "target"]
print(df.shape)
```

```python
tsne = manifold.TSNE(n_components=2, init='pca', random_state=2020)
data_tsne = tsne.fit_transform(df[use_col].values)
```

```python
plt.figure(figsize=(12,8))

plt.plot(data_tsne[df["target"]==1][:,0], data_tsne[df["target"]==1][:,1], '.b', MarkerSize=12, label="class 1")
plt.plot(data_tsne[df["target"]==2][:,0], data_tsne[df["target"]==2][:,1], '.r', MarkerSize=12, label="class 2")
plt.plot(data_tsne[df["target"]==3][:,0], data_tsne[df["target"]==3][:,1], '.g', MarkerSize=12, label="class 3")
plt.plot(data_tsne[df["target"]==12][:,0], data_tsne[df["target"]==12][:,1], '.k', MarkerSize=12, label="class 1&2 mixup")

plt.grid()
plt.legend(fontsize=16)
plt.title("t-SNE visualization on the images (+mixup)", fontsize=20)
```

### Cutmix sequence

```python
df_cutmix_seq =pd.read_csv("../features/cutmix_seq_ae_features.csv")
df = pd.concat([df_train, df_test, df_cutmix_seq], axis=0)
use_col = [col for col in df.columns if col != "target"]
print(df.shape)
```

```python
tsne = manifold.TSNE(n_components=2, init='pca', random_state=2020)
data_tsne = tsne.fit_transform(df[use_col].values)
```

```python
plt.figure(figsize=(12,8))

plt.plot(data_tsne[df["target"]==1][:,0], data_tsne[df["target"]==1][:,1], '.b', MarkerSize=12, label="class 1")
plt.plot(data_tsne[df["target"]==2][:,0], data_tsne[df["target"]==2][:,1], '.r', MarkerSize=12, label="class 2")
plt.plot(data_tsne[df["target"]==3][:,0], data_tsne[df["target"]==3][:,1], '.g', MarkerSize=12, label="class 3")
plt.plot(data_tsne[df["target"]==12][:11][:,0], data_tsne[df["target"]==12][:11][:,1], '.k', MarkerSize=12, label="class 1&2 cutmix")
for i, loc in enumerate(data_tsne[df["target"]==12][:11]):
    plt.text(loc[0]+3, loc[1]+3, "-"+str(i)+"-", fontsize=18)
    
plt.grid()
plt.legend(fontsize=16)
plt.title("t-SNE visualization on the images (+cutmix seq)", fontsize=20)
```
